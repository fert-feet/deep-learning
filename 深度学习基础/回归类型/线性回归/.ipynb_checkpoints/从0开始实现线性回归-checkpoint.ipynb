从 0 实现数据流水线、模型、损失函数和小批量随机梯度下降优化器

```python
%matplotlib inline
import random
import torch
from d2l import torch as d2l
```

### 人造数据集

根据带有噪声的线性模型构造一个人造数据集，使用线性模型参数 $w = [2, -3.4]^T、b = 4.2$ 和噪声项 $\epsilon$ 生成数据集及其标签

```python
def synthetic_data(w, b, num_examples):
    """ 生成 y = Xw + b + noise """
    X = torch.normal(0, 1, (num_examples, len(w))) #从正态分布 N(0,1) 中提取一个 num_example s X len(w) 的矩阵
    y = torch.matmul(X, w) + b #matmul 相当于 mm
    y += torch.normal(0, 0.01, y.shape) #随机噪音
    return X, y.reshape((-1, 1)) # -1指的是让代码自行判断行数，但保持列数为 1， 即转为列向量

true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)
```

`features` 每一行都包含一个二维数据样本，`labels`每一行都包含一维标签值（标量），前者就是 $x$,后者就是 $y$

```python
print('features:',features[0],'\nlabels:',labels[0])
```

```python
d2l.set_figsize()
d2l.plt.scatter(features[:, 1].detach().numpy(),labels.detach().numpy(), 1) #后面的 1 代表点的大小，[:, 1]特征代表第二个维度和标签的关系
```

### 读取小批量

定义 `data_iter`，输入为批量大小、特征矩阵和标签向量，生成大小为 `batch_size` 的小批量

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples)) #列表为从 0 到样本数-1
    random.shuffle(indices) #用于将列表打乱顺序
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(indices[i:min(i + batch_size, num_examples)]) #每次拿 batch_size 个元素
        yield features[batch_indices], labels[batch_indices] #yield 不会结束循环，张量可以用张量（索引）取，即列表可以用列表取元素

batch_size = 10

for X, y in data_iter(batch_size, features, labels):
    print(X, '\n', y)
    break
```

### 定义模型

定义初始化模型参数

这里的 `w` 和 `b` 是真正需要预测的，上面的 `true_w` 和 `true_b` 指的是正确的数据，是用来生成数据集的，正常情况下是不知道这两个数据而只知道数据集的

```python
w = torch.normal(0, 0.01, size = (2, 1), requires_grad = True) #加权，这里的 w
b = torch.zeros(1, requires_grad = True) #偏差
```

模型定义

```python
def linreg(X, w, b):
    """ 线性回归模型 """
    return torch.matmul(X, w) + b
```

### 损失函数定义

这里使用均方误差：$\ell = {\frac 1 2 {(y - \hat y)}}^2$

```python
def squared_loss(y_hat, y):
    return(y_hat - y.reshape(y_hat.shape))**2 / 2
```

### 定义优化算法

使用小批量随机梯度下降：$w_t = w_{t-1} - \eta {\frac {\partial\ell} {\partial w_{t-1}}}$

需要注意，这里是更新数据的函数，不要求导数，在 `sgd` 函数之前会有一个求导过程

```python
def sgd(params, lr, batch_size): # Stochastic Gradient Descent 随机梯度下降
    with torch.no_grad():
        for param in params:
            param -= lr * param.grad / batch_size # lr:学习率，除 batch_size 控制更新幅度
            param.grad.zero_()
```

### 训练过程

实际上就是让损失降到最小（即真实值和预测值最小），以此来寻找最佳的 `w`，`b`，过程中不断优化 `w`，`b`

```python
lr = 0.03# 学习率
num_epochs = 3 # 迭代三次
net = linreg # 模型选择
loss = squared_loss # 损失选择
for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels): # 每次拿一个批量的 'x', 'y'
        l = loss(net(X, w, b), y) # net(X, w, b) 是预测值，l 是 'x' 和 'y' 的小批量损失
        l.sum().backward() # 'l' 是一个列向量，要算总和变为标量再求导
        sgd([w, b], lr, batch_size) # 对 'w', 'b' 进行更新
    with torch.no_grad():
        train_1 = loss(net(features, w, b), labels) # 第 i 次迭代后的 loss
        print(f'epoch {(epoch + 1)}, loss {float(train_1.mean()):f}') # :f 是以 float 形式展示
```

### 总结

整个实现过程可以化为以下步骤：
1. 构造数据集，即生成 $X,y$，其中 $X$ 和 $w$ 维度相同，即 `shape(X) = [num_example, len(w)]`，才能保证可乘
2. 定义小批量读取函数，不能在整个数据集上训练，花费很大，通过多次小批量训练，每次取 `batch_size` 个数据
3. 定义损失函数，最后也是在损失函数上对 $w$ 和 $b$ 求导
4. 定义优化方法，使用优化方法不断更新 $w$ 和 $b$
5. 训练过程，定义学习率和迭代次数，对需要更新的数据不断迭代（$w, b$），并输出每次迭代后的 `loss`

**其实就是不断寻找最优 $w, b$ 的过程**
